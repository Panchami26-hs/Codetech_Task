```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

spark = SparkSession.builder.appName("BigDataAnalysis").getOrCreate()
```

```python
data = spark.read.csv("netflix_titles.csv", header=True, inferSchema=True)
```

```python
data.show(5)
```

```python
print("Total rows:", data.count())
```

```python
print(data.columns)
```

```python
# Movies vs TV Shows
data.groupBy("type").count().show()

# Top countries
data.groupBy("country").count().orderBy("count", ascending=False).show(10)

# Movies after 2015
from pyspark.sql.functions import col
data.filter((col("type") == "Movie") & (col("release_year") > 2015)).show(5)
```
